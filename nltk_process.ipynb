```python
import nltk
from datasets_preprocessing import get_part_of_speech
from Text_Cnn import Vocab
```

```python
sentence = "here 's no one to restrain oldman from making said choices"
words = nltk.word_tokenize(sentence)
print(words)
words_tag = nltk.pos_tag(words)
print(words_tag[9][1])
```

```python
not_changed = 0
total_num = 0
for w in Vocab:
    words[9] = w
    total_num += 1
    if nltk.pos_tag(words)[9][1] == "JJ":
        not_changed += 1
    if total_num%100 == 0:
        print(not_changed/total_num)
    if total_num == 100000:
        break
print(not_changed/total_num)
```
